```python
#Dataset - Area vs Prices
#Dataset - https://raw.githubusercontent.com/ameenmanna8824/DATASETS/main/areavsprices.csv
#Area - Sqft
#Prices - Rupees
```

```python
#1.Take the data and create dataframe
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/ameenmanna8824/DATASETS/main/areavsprices.csv')
df
```

```python
#2.PRE PROCESSING - IT IS OPTIONAL ,HENCE WE ARE NOT PERFORMING IT
```

```python
#3.DATA VISUALISATION
import matplotlib.pyplot as plt
plt.scatter(df['Area'],df['Prices'])
plt.title('AREA vs PRICES')
plt.xlabel('Area')
plt.ylabel('Prices')
```

```python
#Area - Input
#Prices - Output
```

```python
#4.DIVIDE INTO INPUT and OUTPUT
#Input(x)- is always 2 dimensional array
#Output(y)-is always 1 dimensional array

#var.iloc[row slicing,column slicing].By wrirting : in row slicing part , all the rows are selected
#In the Column slicing part , if there is a ':', then the array is 2 dimensional
x = df.iloc[0:6,0:1].values #df.iloc[:,0:1].values 
x # .values convert the dataframe into an array
```

```python
#In the column slicing part , if there is no : , then the array is 1 dimensional
y = df.iloc[:,1].values
y
```

```python
#5.Train and test variables
#Here due to limited data , we are not performing step no 5
```

```python
#6.NORMALISATION or SCALING
#It is only to be performed for Multivariate dataset,here our dataset is univariate,hence not required
```

```python
#7.Run a classifier,REGRESSOR or clusterer
#sklearn.linear_model - package,LinearRegression - Library
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#8.FIT the MODEL(mapping/plotting the inputs with the outputs)
#LinearRegression.fit(x,y)
model.fit(x,y) #We are plotting the values of x and y in the LinearRegression Library.
```

```python
#9.PREDICT THE OUTPUT
y_pred = model.predict(x) #Using the input values of Area , We predict the output Prices
y_pred #PREDICTED OUTPUT VALUES
```

```python
y #ACTUAL OUTPUT VALUES
```

```python
#CONCLUSION : Wehave to compare both y_pred and y.
#SO when we compare the corresponding values of Y-pred and y , we come to know that ther eis huge difference.
#This huge difference does not mean that our model has predicted wrong.
#It only means that , our model is NOT LINEAR/LESS LINEAR.
#LINEARITY of a MODEL depends on the NATURE of the data as well as the size of the data.
```

```python
#INDIVIDUAL PREDICTION
#I want to know the price of 2000 sqft
model.predict([[]])2000
```

```python
#CROSS VERIFY
#y = mx + C
#m - slope
#C - y-intercept/constant
#y - dependant variable
#x - independant variable 
```

```python
#To find out the value of slope -m
m = model.coef_
m
```

```python
#To find out the value of y-intercept (C)
C = model.intercept_
C
```

```python
#Substitute in y = mx + C
m * 2000 + C
```

```python
#VISUALISATION - FOR THE BEST FIT LINE
plt.scatter(x,y) #ACTUAL VALUES
plt.plot(x,y_pred,color = 'orange') #PREDICTED VALUES
plt.title('BEST FIT LINE')
```
