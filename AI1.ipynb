```python
#1.Read the IMAGE

import cv2 #we are importing the opencv/computer vision library
img = cv2.imread('pic.jpg') #we are reading the image and storing it inside the img variable
cv2.imshow('OUTPUT1',img) #Here we are displaying the image

cv2.waitKey(3000)#cv2.waitKey(0) makes the image stay forever
#cv2.waitKey(3000) makes the image stay onlt for 3 seconds
cv2.destroyAllWindows()
```

```python
#2.CREATE a duplicate image of any extension

import cv2
img = cv2.imread('pic.jpg') #reading the image
cv2.imshow('OUTPUT1',img)#displaying the image

cv2.imwrite('ameen.png',img)#create a new image with a new extension

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#3.Reading the information of any particular image

import cv2
img = cv2.imread('pic.jpg')
print(img.shape)

#OUTPUT: (1200, 1920, 3)
#1200 is the Height of the image in pixels
#1920 is the Width of the image in pixels
#3 is the channel value/channel number(RGB)/depth
```

```python
#4.GRAY SCALE IMAGE (BLACK AND WHITE IMAGE)

import cv2
img = cv2.imread('pic.jpg')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

cv2.imshow('ORIGINAL IMAGE',img)#displays the original image
cv2.imshow('GRAY SCALE IMAGE',gray)#displays the gray scale image

cv2.waitKey(0)
cv2.destroyAllWindows()
    

```

```python
#GRAY SCALE SHORT PROGRAM

import cv2
img = cv2.imread('pic.jpg',0) #0 here converts my image into grayscale

cv2.imshow('GRAY SCALE IMAGE',img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
