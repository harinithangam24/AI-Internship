```python
#MATPLOTLIB - VISUALISATION(GRAPHS and REPRESENTATIONS)
#1.LINE GRAPH/PLOT using LISTS
import matplotlib.pyplot as plt
a = [1,2,3,4]
b = [5,6,7,8]
#plt.plot(x-axis,y-axis)
plt.plot(a,b,color = 'orange',marker = '*')
plt.title('LINE GRAPH/PLOT')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
```

```python
#2.LINE GRAPH/PLOT using NUMPY array
x = np.array([1,2,3,4])
y = np.array([5,6,7,8])
plt.plot(x,y,color = 'orange',marker = 'v')
plt.title('LINE GRAPH/PLOT')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
```

```python
#3.SCATTER PLOT
a = [1,2,3,4]
b = [5,6,7,8]
#plt.scatter(x-axis,y-axis)
plt.scatter(a,b,color = ['orange','red','gold','lime'])
plt.title('SCATTER PLOT')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
```

```python
#4.BAR GRAPH/PLOT
import matplotlib.pyplot as plt
names = ['Ashik','Sam','Chris','Gopal','Rahul']
weight = [78,86,67,77,81]
#plt.bar(x-axis,y-axis)
plt.bar(names,weight,color = ['Orange','red','gold','lime','deepskyblue'])
plt.xlabel('Names')

plt.ylabel('Weight')
```
